# Trabalhando com datas

```python
# Importando a biblioteca
import pandas as pd
```

```python
# Leitura dos arquivos
df1 = pd.read_excel("C:/Projectos/datasets/Aracaju.xlsx")
df2 = pd.read_excel("C:/Projectos/datasets/Fortaleza.xlsx")
df3 = pd.read_excel("C:/Projectos/datasets/Natal.xlsx")
df4 = pd.read_excel("C:/Projectos/datasets/Recife.xlsx")
df5 = pd.read_excel("C:/Projectos/datasets/Salvador.xlsx")
```

```python
# Juntando todos os arquivos
df = pd.concat([df1,df2,df3,df4,df5])
```

```python
# Exibindo as 5 primeiras linhas
df.head()
```

```python
# Criando a coluna de receita
df["Receita"]=df["Vendas"].mul(df["Qtde"])
```

```python
# Transformando a coluna de data em tipo inteiro
df["Data"] = df["Data"].astype("Int64")
```

```python
# Verificando o tipo de dado de cada coluna
df.dtypes
```

```python
# Tranformando coluna de data em data
df["Data"] = pd.to_datetime(df["Data"])
```

```python
df.dtypes
```

```python
# Agrupamento por ano
df.groupby(df["Data"].dt.year)["Receita"].sum()
```

```python
# Criando uma nova coluna com o ano
df["Ano_Venda"] = df["Data"].dt.year
```

```python
df.sample(5)
```

```python
# Extraindo o mes e o dia
df["mes_venda"], df["dia_venda"] = (df["Data"].dt.month, df["Data"].dt.day)
```

```python
df.sample(5)
```

```python
# Retornando a data mais antiga
df["Data"].min()
```

```python
# Calculando a diferenca de dias
df["diferenca_dias"] = df["Data"] - df["Data"].min()
```

```python
df.sample(5)
```

```python
# Criando a coluna de trimentre
df["trimestre_de_venda"] = df["Data"].dt.quarter
```

```python
df.sample(5)
```

```python
# Filtrando as vendas de 2019 do mes de marco
vendas_marco_19 = df.loc[(df["Data"].dt.year == 2019) & (df["Data"].dt.month == 3)]
```
