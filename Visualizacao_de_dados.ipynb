# Visualizacao de dados

```python
# Importando a biblioteca
import pandas as pd
```

```python
# Leitura dos arquivos
df1 = pd.read_excel("C:/Projectos/datasets/Aracaju.xlsx")
df2 = pd.read_excel("C:/Projectos/datasets/Fortaleza.xlsx")
df3 = pd.read_excel("C:/Projectos/datasets/Natal.xlsx")
df4 = pd.read_excel("C:/Projectos/datasets/Recife.xlsx")
df5 = pd.read_excel("C:/Projectos/datasets/Salvador.xlsx")
```

```python
# Juntando todos os arquivos
df = pd.concat([df1,df2,df3,df4,df5])
```

```python
# Criando a coluna de receita
df["Receita"]=df["Vendas"].mul(df["Qtde"])
```

```python
df["LojaID"].value_counts(ascending=False)
```

```python
# Grafico de Barras
df["LojaID"].value_counts(ascending=False).plot.bar()
```

```python
# Grafico de barras horizontais
df["LojaID"].value_counts().plot.barh()
```

```python
# Grafico de barras horizontais
df["LojaID"].value_counts(ascending=True).plot.barh()
```

```python
# Grafico de Pizza
df.groupby(df["Data"].dt.year)["Receita"].sum().plot.pie()
```

```python
# Total vendas por cidade
df["Cidade"].value_counts()
```

```python
# Adicionando un titulo e alterand o nome dos eixos
import matplotlib.pyplot as plt
df["Cidade"].value_counts().plot.bar(title="Total vendas por Cidade")
plt.xlabel("Cidade")
plt.ylabel("Total Vendas");
```

```python
# Alterando a cor
df["Cidade"].value_counts().plot.bar(title="Total vendas por Cidade", color="red")
plt.xlabel("Cidade")
plt.ylabel("Total Vendas");
```

```python
# Alterando o estilo
plt.style.use("ggplot")
```

```python
# Extraindo o mes e o dia
df["mes_venda"], df["dia_venda"] = (df["Data"].dt.month, df["Data"].dt.day)
```

```python
df.groupby(df["mes_venda"])["Qtde"].sum().plot()
plt.xlabel("Mes")
plt.ylabel("Total Produtos Vendidos")
plt.legend();
```

```python
df.groupby(df["mes_venda"])["Qtde"].sum()
```

```python
# Criando uma nova coluna com o ano
df["Ano_Venda"] = df["Data"].dt.year
```

```python
# Seleccionando apenas as vendas de 2019
df_2019 = df[df["Ano_Venda"] == 2019]
```

```python
# Total produtos vendidos por mes
df_2019.groupby(df_2019["mes_venda"])["Qtde"].sum().plot(marker = "v")
plt.xlabel("Mes")
plt.ylabel("Total Produtos Vendidos");
plt.legend()
```

```python
# Histograma
plt.hist(df["Qtde"], color= "magenta");
```

```python
plt.scatter(x=df_2019["dia_venda"], y = df_2019["Receita"]);
```

```python
# Salvando em png
df_2019.groupby(df_2019["mes_venda"])["Qtde"].sum().plot(marker = "v")
plt.title("Quantidade de produtos vendidos x mes")
plt.xlabel("Mes")
plt.ylabel("Total Produtos Vendidos");
plt.legend()
plt.savefig("grafico QTDE x MES.png")
```
